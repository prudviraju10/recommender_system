```python
df_exploded = pd.get_dummies(df_exploded, columns=['Genres'])
```

```python
for col in df_exploded.columns[11:]:
    df_exploded[col] = df_exploded[col].astype(int)
```

```python
df_exploded.groupby('title')['Rating'].mean().sort_values(ascending=False).head(5)
```

```python
df_exploded["avg_rating_mve"] = df_exploded.groupby("MovieID")["Rating"].transform("mean")
df_exploded["avg_rating_user"] = df_exploded.groupby("UserID")["Rating"].transform("mean")
df_exploded["count_rating_mve"] = df_exploded.groupby("MovieID")["Rating"].transform("count")
df_exploded["count_rating_user"] = df_exploded.groupby("UserID")["Rating"].transform("count")
```

```python
df_exploded["Zip-code"] = df_exploded.groupby("Zip-code")["Rating"].transform("mean")
```

```python
pd.options.display.max_columns = None
```

```python
df_exploded.head(2)
```

```python
df_exploded.isna().sum()
```

```python
df_exploded.info()
```

```python
X = df_exploded.drop(["Rating","UserID","MovieID","title"], axis=1)
y = df_exploded["Rating"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
mse(y_test, y_pred)**0.5
```

```python
df_ratings = pd.read_csv("../data/ratings.csv")
```

```python
df_ratings_cmfrec = df_ratings[["UserID","MovieID","Rating"]]
df_ratings_cmfrec.columns = ['UserId', 'ItemId', 'Rating'] 
df_ratings_cmfrec.head()
```

```python
from cmfrec import CMF

model = CMF(method="als", k=6, lambda_=0.1, user_bias=False, item_bias=False, verbose=False)
model.fit(df_ratings_cmfrec)
```

```python
model.A_.shape, model.B_.shape
```

```python
A = model.A_
A[:,0]
```

```python
# sns.scatterplot(x=A[:,0], y=A[:,1])
```

```python
B = model.B_
# sns.scatterplot(x=B[:,0], y=B[:,1])
```

```python
df_ratings_cmfrec.Rating.mean(), model.glob_mean_
```

```python
# ratings given by user 1 to movie 1 is
model.predict([0], [0])
```

```python
model.predict(5,6)
```

```python
print(np.dot(A[0,:], B[0,:])+ model.glob_mean_)
print(np.dot(A[1,:], B[20,:])+ model.glob_mean_)
```

```python
rm__ = np.dot(model.A_, model.B_.T) + model.glob_mean_
mse(df_mat.values[df_mat > 0], rm__[df_mat > 0])**0.5
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
mape(df_mat.values[df_mat > 0], rm__[df_mat > 0])
```
